```python
import networkx as nx
import matplotlib.pyplot as plt
from operator import itemgetter
```

```python
G = nx.karate_club_graph()
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
```

```python
bet = nx.betweenness_centrality(G)
sorted(bet.items(), key=itemgetter(1), reverse=True)
```

Najwiksze beetweenness centrality ma weze 0, warto: 0.43763528138528146, czyli okoo 44%

```python
nodes = G.nodes()
values = [bet.get(node) for node in nodes]
values
```

```python
ec = nx.draw_networkx_edges(G, pos, alpha=0.2)
nc = nx.draw_networkx_nodes(G, pos, nodelist=nodes, node_color=values, cmap=plt.cm.plasma_r)

plt.colorbar(nc)
plt.axis('off')
plt.show()
```
